Uvozimo vse potrebne module in definiramo strukturo vhodnih podatkov za funkcijo primerjaj2. Ekipi morata biti iz seznama possible_teams, statistika, ki jo išemo mora biti iz seznama possible_stats in leto za katero poizvedujemo mora biti iz intervala [1990,2023]. Ker nekatere ekipe niso nastopale v vseh letih in posledično za njih ne obstaja tabela, potrebujemo try/except stavek s katerim ovijemo glavno vsebino funkcije, da ne bomo imeli nepričakovanih izhodov za neskladne podatke. Prva stvar, ki jo preverimo v funkciji je da vhodni podatki ustrezajo prej predpisanim. Nato shranimo v ekipa1_data in ekipa2_data podatke iz slovarja datagramov in vzamemo samo stolpec, po katerem poizvedujemo (stat). Podatke o obeh ekipah združimo v spremenljivko combined za lažjo obdelavo. Glede na tip statistike imamo 3 različne predstavitve podatkov, ker se tudi podatki sami razlikujejo po vsebini jih rabimo obravnavati posebej.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

possible_stats = ['Opp', 'Tm', 'W', 'L', 'Streak', 'Attend.',] 
possible_years = list(range(1990, 2024))
possible_teams = ['ATL', 'BOS', 'SAS', 'NJN', 'CHA', 'CHI', 'CLE', 
                   'DAL', 'DEN', 'DET', 'GSW', 'HOU', 'IND', 'LAC', 
                   'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOH', 'NYK', 
                   'OKC', 'ORL', 'PHI', 'PHO', 'POR', 'SAC', 'TOR',
                   'UTA', 'WAS']

def primerjaj2 (ekipa1, ekipa2, leto, stat):
    try:
        if stat in possible_stats and leto in possible_years and ekipa1 in possible_teams and ekipa2 in possible_teams:
            ekipa1_data = dfs[ekipa1][leto]
            ekipa2_data = dfs[ekipa2][leto]
            
            ekipa1_data = ekipa1_data[[stat]].copy()
            ekipa1_data['Team'] = ekipa1
            ekipa1_data['Game'] = range(1, len(ekipa1_data)+1)
    
            ekipa2_data = ekipa2_data[[stat]].copy()
            ekipa2_data['Team'] = ekipa2
            ekipa2_data['Game'] = range(1, len(ekipa2_data)+1)
    
            combined = pd.concat([ekipa1_data, ekipa2_data])
            
            if stat in ['Tm', 'Opp', 'Attend.']:
                plt.figure(figsize=(14,6))
                sns.lineplot(data=combined, x='Game', y=stat, hue='Team', marker="o")
                mean1 = round(dfs[ekipa1][leto][stat].dropna().astype(int).mean(), 2)
                mean2 = round(dfs[ekipa2][leto][stat].dropna().astype(int).mean(), 2)
                
                if stat == 'Tm': 
                    stat = 'Dosežene točke'
                elif stat == 'Opp': 
                    stat = 'Prejete točke'
                elif stat == 'Attend.':
                    stat = 'Število gledalcev na tekmah'
                plt.ylabel(stat)
                plt.title(f'{stat} skozi sezono {leto} za {ekipa1} in {ekipa2}')
                plt.grid(True)
                plt.show()
                
                print("Povprecje {} ekipe {} v letu {}/{} je bilo: {}". format(stat, ekipa1, leto, leto+1, mean1))
                print("Povprecje {} ekipe {} v letu {}/{} je bilo: {}". format(stat, ekipa2, leto, leto+1, mean2))
            elif stat in ['W', 'L']:
                st1 = ekipa1_data[stat].iloc[-1]
                st2 = ekipa2_data[stat].iloc[-1]
                
                if stat == 'W':
                    stat = 'Zmage'
                else:
                    stat = 'Porazi'
                
                plt.figure(figsize=(8,6))
                plt.bar([ekipa1, ekipa2], [st1, st2], color=['blue', 'green']) 
                plt.xlabel("Ekipi")
                plt.ylabel(stat)
                plt.title("{} v sezoni {}/{} za ekipi {} in {}.".format(stat, leto, leto+1, ekipa1, ekipa2))
                plt.grid(axis='y')
                plt.show()
            else:
                w_streaks1 = ekipa1_data['Streak'][ekipa1_data['Streak'].str.startswith('W')]
                l_streaks1 = ekipa1_data['Streak'][ekipa1_data['Streak'].str.startswith('L')]
                w_streaks_numbers1 = w_streaks1.str.extract(r'W\s*(\d+)').astype(int)
                max_w_streak1 = w_streaks_numbers1.max()[0]
                l_streaks_numbers1 = l_streaks1.str.extract(r'L\s*(\d+)').astype(int)
                max_l_streak1 = l_streaks_numbers1.max()[0]

                w_streaks2 = ekipa2_data['Streak'][ekipa2_data['Streak'].str.startswith('W')]
                l_streaks2 = ekipa2_data['Streak'][ekipa2_data['Streak'].str.startswith('L')]
                w_streaks_numbers2 = w_streaks2.str.extract(r'W\s*(\d+)').astype(int)
                max_w_streak2 = w_streaks_numbers2.max()[0]
                l_streaks_numbers2 = l_streaks2.str.extract(r'L\s*(\d+)').astype(int)
                max_l_streak2 = l_streaks_numbers2.max()[0]
    
                print("Najvecji zmagovalni niz ekipe {} je bil: {}".format(ekipa1, max_w_streak1))
                print("Najvecji niz porazov ekipe {} je bil: {}".format(ekipa1, max_l_streak1))
                print("Najvecji zmagovalni niz ekipe {} je bil: {}".format(ekipa2, max_w_streak2))
                print("Najvecji niz porazov ekipe {} je bil: {}".format(ekipa2, max_l_streak2))
        else:
            if stat not in possible_stats:
                print("Napačno ime statistike, izberi enega izmed naslednjih: 'Opp', 'Tm', 'W', 'L', 'Streak', 'Attend.' in poskusi še enkrat.")
            elif leto not in possible_years:
                print("Neveljavno leto, izberi leto iz intervala [1990, 2023]")
            elif ekipa1 not in possible_teams or ekipa2 not in possible_teams: 
                print("Ena ali obe izmed ekip niso del veljavnih, izberi ju iz seznama: {}".format(possible_teams))
            return
    except Exception as e:
        print(e) 
        print("Ekipa {} ali {} ni nastopala v letu {}".format(ekipa1, ekipa2, leto))
        return
```

Primerjaj1 je funkcija, ki nam za ekipo in specifiran stat v časovnem obdobju od leto1 do vključno leto2 vrne primerjavo po letih. Za zmage/poraze nariše graf, kjer so na x-osi leta in na y-osi zmage/porazi v tistem letu. Za tm, opp, attend. pa nam vrne graf s povprečnimi vrednostmi za vsako leto. Ker tukaj ne operiramo več z dataframe-i ampak s podatki, ki jih iz df izluščimo in shranimo v seznam, je najlažje risati z matplotlib knjižnico in ne s seaborn.

```python
def primerjaj1(ekipa, leto1, leto2, stat):
    try:
        if stat in possible_stats and leto1 <= leto2 and leto1 in possible_years and leto2 in possible_years and ekipa in possible_teams: 
            
            #za W/L, vsako leto iz intervala [leto1, leto2] zabelezimo kolko zmag in kolko porazov je imela ekipa 
            if stat in ['W', 'L']:
                tabela = []
                leta = list(range(leto1, leto2))
                
                for leto in leta: 
                    ekipa_data = dfs[ekipa][leto]
                    st = ekipa_data[stat].iloc[-1]  # število zmag/porazov
                    tabela.append(st) 
                
                plt.figure(figsize=(14,10)) 
                plt.plot(leta, tabela, '-o')
                plt.xlabel("Leto")
                plt.ylabel(stat)
                plt.title("Število {} v sezonah od {} do {} za ekipo {}.".format(stat, leto1, leto2, ekipa))
                plt.grid(True)
                
                for i in range(len(leta)):
                    plt.annotate(f'{tabela[i]}', (leta[i], tabela[i]), textcoords="offset points", xytext=(0, 8), ha='center', color = 'green')
                
                plt.xticks(leta, rotation=45)
                plt.tight_layout()
                plt.ylim(0, 82)
                plt.show()
            
            #za Tm, Opp, Attend. bomo vsako leto vzeli povprečja
            if stat in ['Tm', 'Opp', 'Attend.']:
                tabela = [] 
                leta = list(range(leto1, leto2))

                for leto in leta:
                    mean = round(dfs[ekipa][leto][stat].dropna().astype(int).mean(), 1)
                    tabela.append(mean)

                plt.figure(figsize=(14,10))
                plt.plot(leta, tabela, '-o')
                plt.xlabel("Leto")
                
                if stat == 'Tm':
                    stat = "Dosežene točke"
                elif stat == 'Opp':
                    stat = "Prejete točke"
                else:
                    stat = "Gledalci"
                
                plt.ylabel(stat)
                plt.title("Povprečno število {} v sezonah od {} do {} za ekipo {}.".format(stat, leto1, leto2, ekipa))
                plt.grid(True)
                
                for i in range(len(leta)):
                    plt.annotate(f'{tabela[i]}', (leta[i], tabela[i]), textcoords="offset points", xytext=(0, 8), ha='center', color = 'green')
                
                plt.xticks(leta, rotation=45)
                plt.tight_layout()
                plt.show()

            if stat == "Streak":
                tabela = []
                leta = list(range(leto1, leto2))
                
                for leto in leta: 
                    ekipa_data = dfs[ekipa][leto]
                    w_streaks = ekipa_data['Streak'][ekipa_data['Streak'].str.startswith('W')]
                    w_streaks_numbers = w_streaks.str.extract(r'W\s*(\d+)').astype(int)
                    max_w_streak = w_streaks_numbers.max()[0]
                    tabela.append(int(max_w_streak))  
                
                plt.figure(figsize=(14,10)) 
                plt.plot(leta, tabela, '-o')
                plt.xlabel("Leto")
                plt.ylabel(stat)
                plt.title("Najdaljši niz zmag v sezonah od {} do {} za ekipo {}.".format(leto1, leto2, ekipa))
                plt.grid(True)
                
                for i in range(len(leta)):
                    plt.annotate(f'{tabela[i]}', (leta[i], tabela[i]), textcoords="offset points", xytext=(0, 8), ha='center', color = 'green')
                
                plt.xticks(leta, rotation=45)
                plt.tight_layout()
                plt.show()
        else:
            if stat not in possible_stats:
                print("Napačno ime statistike, izberi enega izmed naslednjih: 'Opp', 'Tm', 'W', 'L', 'Streak', 'Attend.' in poskusi še enkrat.")
            elif leto1 not in possible_years or leto2 not in possible_years:
                print("Neveljavno leto, izberi leto iz intervala [1990, 2023]")
            elif ekipa not in possible_teams: 
                print("Ekipa ni del veljavnih, izberi jo iz seznama: {}".format(possible_teams))
            return
    except Exception as e:
        print(e)
        print("Ekipa {} ni nastopala v letu {}".format(ekipa, leto))
        return
```

Ker hočem naredit še eno funkcijo, ki vzame 2 ekipi, interval let in stat, ter ju primerja na istem grafu, je najlažje da recikliram del kode iz primerjaj1, in sicer del, ki mi shrani statistične podatke, v novi funkciji statistika.

```python
def statistika(ekipa, leto1, leto2, stat):
    if stat not in possible_stats or leto1 not in possible_years or leto2 not in possible_years or ekipa not in possible_teams:
        return None
    
    tabela = []
    leta = list(range(leto1, leto2))

    if stat in ['W', 'L']:
        for leto in leta: 
            try:
                st = dfs[ekipa][leto][stat].iloc[-1]
            except (KeyError, TypeError):
                st = None
            tabela.append(st)
    elif stat in ['Tm', 'Opp', 'Attend.']:
        for leto in leta:
            try:
                mean = round(dfs[ekipa][leto][stat].dropna().astype(int).mean(), 1)
            except (KeyError, TypeError):
                mean = None
            tabela.append(mean)
    elif stat == "Streak":
        for leto in leta: 
            try:
                w_streaks = dfs[ekipa][leto]['Streak'][dfs[ekipa][leto]['Streak'].str.startswith('W')]
                w_streaks_numbers = w_streaks.str.extract(r'W\s*(\d+)').astype(int)
                max_w_streak = w_streaks_numbers.max()[0]
            except (KeyError, TypeError):
                max_w_streak = 0
            tabela.append(int(max_w_streak))
    return leta, tabela
```

Sedaj definiramo primerjaj1_2, ki sprejme 2 ekipi, interval let ter stat in vrne primerjavo med njima čez leta.

```python
def primerjaj1_2(ekipa1, ekipa2, leto1, leto2, stat):
    try:
        data1 = statistika(ekipa1, leto1, leto2, stat)
        data2 = statistika(ekipa2, leto1, leto2, stat)
        if data1 is None and data2 is None:
            print("Napaka pri pridobivanju statistike za {} in {} in leta [{}, {}].".format(ekipa1, ekipa2, leto1, leto2))
            return
        elif data1 is None and not (data2 is None):
            print("Napaka pri pridobivanju statistike za {} in leta [{}, {}].".format(ekipa1, leto1, leto2))
            return
        elif data2 is None and not (data1 is None):
            print("Napaka pri pridobivanju statistike za {} in leta [{}, {}].".format(ekipa2, leto1, leto2))
            return
        leta1, tabela1 = data1
        leta2, tabela2 = data2
        plt.figure(figsize=(14,10))
        plt.plot(leta1, tabela1, '-o', label=ekipa1, color='blue')
        plt.plot(leta2, tabela2, '-o', label=ekipa2, color='red')
        plt.xlabel("Leto")

        y_label = stat
        if stat == 'Tm':
            y_label = "Dosežene točke"
        elif stat == 'Opp':
            y_label = "Prejete točke"
        elif stat == 'Attend.':
            y_label = "Gledalci"

        plt.ylabel(y_label)
        plt.title(f"{y_label} v sezonah od {leto1} do {leto2} za ekipe {ekipa1} in {ekipa2}")
        plt.grid(True)
        
        for i in range(len(leta1)):
            plt.annotate(f'{tabela1[i]}', (leta1[i], tabela1[i]), textcoords="offset points", xytext=(0, 8), ha='center', color='blue')
        for i in range(len(leta2)):
            plt.annotate(f'{tabela2[i]}', (leta2[i], tabela2[i]), textcoords="offset points", xytext=(0, 8), ha='center', color='red')

        plt.xticks(leta1, rotation=45)
        plt.tight_layout()
        plt.legend()
        plt.show()
        return
    except Exception as e:
        print(e)
        print("Napaka pri primerjavi.")
        return
```

Smiselno se mi zdi še dodati primerjavo vseh ekip po določeni statistiki v enem letu. Definiral bom funkcijo primerjaj_leto(leto, stat), ki vrne primerjavo vseh ekip po statistiki "stat" v letu "leto".

```python
def primerjaj_leto(leto, stat): 
    try:
        ekipe = []
        vrednosti = [] 
        for ekipa in possible_teams:
            data = statistika(ekipa, leto, leto+1, stat)
            if data is None:
                continue 
            l, tabela = data
            if len(tabela) == 0 or tabela[0] is None:
                continue  
            ekipe.append(ekipa)
            vrednosti.append(tabela[0])
    
        skupaj = list(zip(ekipe, vrednosti))
        
        if not skupaj:  
            print(f"Ni podatkov za leto {leto} in statistiko '{stat}'.")
            return
        
        skupaj.sort(key=lambda x: x[1])
        ekipe_sorted, vrednosti_sorted = zip(*skupaj)

        if stat == "Streak":
            vrednosti_sorted = [int(v) for v in vrednosti_sorted]
        
        plt.figure(figsize=(14,10))
        plt.bar(ekipe_sorted, vrednosti_sorted)
        plt.xticks(rotation=45, ha="right")
        plt.ylabel(stat)
        plt.title(f"Primerjava ekip v {leto} po statistiki '{stat}'")
        plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(e)
        print("Napaka pri primerjavi.")
```
