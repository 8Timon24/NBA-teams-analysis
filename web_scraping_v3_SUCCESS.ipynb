```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import os

kratice = ['ATL', 'BOS', 'SAS', 'NJN', 'CHA', 'CHI', 'CLE', 
           'DAL', 'DEN', 'DET', 'GSW', 'HOU', 'IND', 'LAC', 
           'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOH', 'NYK', 
           'OKC', 'ORL', 'PHI', 'PHO', 'POR', 'SAC', 'TOR',
           'UTA', 'WAS']
leta = list(range(1993, 2025))
url_start = "https://www.basketball-reference.com/teams/{}/{}_games.html"
base_dir = r"C:\Users\Timon\Desktop\PythonProjects\NBA_teams_through_years\NBA_TEAMS"

options = Options()
options.headless = True  
driver = webdriver.Chrome(options=options)
for leto in leta: 
    for ekipa in kratice:
        try:
            url = url_start.format(ekipa, leto)
            driver.get(url)
            html = driver.page_source
            target_dir = os.path.join(base_dir, ekipa, str(leto))
            os.makedirs(target_dir, exist_ok=True)

            file_path = os.path.join(target_dir, f"{ekipa}_{leto}_games.html")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(html)

            print(f"Shranjeno: {file_path}")
            time.sleep(5)

        except Exception as e:
            print(f"Napaka pri {ekipa}_{leto}: {e}")

driver.quit()
```
