```python
import pandas as pd

url = "https://www.basketball-reference.com/teams/SAS/"

tables = pd.read_html(url)

for table in tables:
    if 'Season' in table.columns:
        sas_table = table
        break

sas_table.to_csv("sas_team_history.csv", index=False)
```

```python
sas_path = sas_path = r"C:\Users\timon\Desktop\PythonProjects\NBA_teams_through_years\sas_team_history.csv"
sas_data = pd.read_csv(sas_path, index_col="Season")
```

```python
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
sas_data.index = sas_data.index.astype(str)
sas_filtered = sas_data.loc[sas_data.index < "2025-26"]
sas_filtered = sas_filtered.reset_index()
sas_filtered = sas_filtered.sort_values("Season")

plt.figure(figsize=(30,10))
plt.title("San Antonio Spurs wins through the years")
plt.ylabel("Wins")
plt.xlabel("Season")
sns.barplot(x="Season", y="W", data=sas_filtered)
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

```python
sas_filtered["Season_num"] = range(len(sas_filtered))
plt.figure(figsize=(25, 6))
plt.title("San Antonio Spurs Wins (with Regression)")
sns.regplot(x="Season_num", y="W", data=sas_filtered, scatter_kws={"color": "blue"}, line_kws={"color": "red"}, ci=None)
plt.xticks(ticks=sas_filtered["Season_num"], labels=sas_filtered["Season"], rotation=45)
plt.ylabel("Wins")
plt.xlabel("Season")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(25, 6))
plt.title("San Antonio Spurs Wins")
plt.ylabel("Wins")
plt.xlabel("Season")

sns.lineplot(x="Season_num", y="W", data=sas_filtered)
plt.xticks(ticks=sas_filtered["Season_num"], labels=sas_filtered["Season"], rotation=45)
plt.tight_layout()
plt.show()
```
