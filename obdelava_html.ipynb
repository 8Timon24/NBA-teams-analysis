Za začetek nastavimo obdobje iz katerega hočemo podatke in od katerih ekip. Vse ekipe niso nujno igrale v vseh letih, ker so se franšize ali preimenovale ali preselile v drugo mesto. S tem se ukvarjamo kasneje.

```python
kratice = ['ATL', 'BOS', 'SAS', 'NJN', 'CHA', 'CHI', 'CLE', 
           'DAL', 'DEN', 'DET', 'GSW', 'HOU', 'IND', 'LAC', 
           'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOH', 'NYK', 
           'OKC', 'ORL', 'PHI', 'PHO', 'POR', 'SAC', 'TOR',
           'UTA', 'WAS']
leta = list(range(1990, 2024))
```

Importamo vse potrebne libraryje, ki jih potrebujemo za obdelavo html strani: pandas, BeautifulSoup, seaborn so glavni, html5lib in StringIO so samo za določene funkcije, ki ju zahtevajo. Začnemo tako da odpremo datoteko in inicializiramo soup, ki parsa ta html. Tukej sem začel z enim primerom da sem vidu, če deluje in kaj rabim sploh naredit.

```python
import pandas as pd 
import seaborn as sns
from bs4 import BeautifulSoup
from io import StringIO
import html5lib

base_dir = r"C:\Users\Timon\Desktop\PythonProjects\NBA_teams_through_years\NBA_TEAMS\{}\{}\{}_{}_games.html"
file_path = r"C:\Users\Timon\Desktop\PythonProjects\NBA_teams_through_years\NBA_TEAMS\ATL\1990\ATL_1990_games.html"

with open(file_path, encoding='utf-8') as f:
    page = f.read()
soup = BeautifulSoup(page, "html.parser")
```

Glavna zanka, iteriramo skozi vse ekipe in leta, ki smo jih v prvem koraku specifirali in jih shranimo v slovar dfs (dataframes), katerega ključ so ekipa in leto (2D slovar je). Če pokličemo: 
dfs[ekipa][leto] -> dobimo tabelo za ekipo 'ekipa' v letu 'leto'. Ker nekatere ekipe v določenih letih niso obstajale, potrebujemo try/except stavek.

```python
dfs = {}
for ekipa in kratice:
    df = {}
    for leto in leta:
        file_tmp = base_dir.format(ekipa, leto, ekipa, leto)
        try:
            with open(file_tmp, encoding='utf-8') as f_tmp:
                page_tmp = f_tmp.read()
            soup = BeautifulSoup(page_tmp, "html.parser")
            for tr in soup.find_all('tr', class_='thead'):
                tr.decompose()
            games_tmp = soup.find(id="games")
            if games_tmp is None:
                print(f"Ni tabele z id='games' za ekipo {ekipa} in leto {leto}")
                continue   
            games_ekipa_leto = pd.read_html(StringIO(str(games_tmp)))[0]
            df[leto] = games_ekipa_leto
        except Exception as e:
            print(f"Ni podatkov za ekipo {ekipa} in leto {leto}: {e}")
            continue
    dfs[ekipa] = df
```
